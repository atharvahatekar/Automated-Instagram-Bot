Problem 1 : Login to your Instagram


```python
def LogIn(username, password):
    try :
        #locating username textbox and sending username
        user_name = wait.until(EC.presence_of_element_located((By.NAME,'username')))
        user_name.send_keys(username)
        #locating password box and sending password
        pwd = driver.find_element('name','password')
        pwd.send_keys(password)
        #locating login button 
        button = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="loginForm"]/div[1]/div[3]/button/div')))
        button.submit()
        #Save Your Login Info? : Not Now
        pop = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/section/main/div/div/div/div/div')))
        pop.click()
        #Turn on Notification? : Not Now
        pop2 = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[3]/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[3]/button[2]')))
        pop2.click()
        
    except TimeoutException :
        print ("Check is login successful or not?")
```

```python
#Login to your Instagram Handle
LogIn(username, password)
```

### Problem 2 : Type for “food” in search bar
Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”  
Note : Make sure to avoid printing hashtags  

```python
def search(toSearch):
    account_names=[]
    #Finding search field
    search_field = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div/a/div')))
    search_field.click()
    #Find the search box and sending the search query
    search_box = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[1]/div/div/input')))
    search_box.send_keys({toSearch})
    time.sleep(5)
    html_data = driver.page_source
    soup=BeautifulSoup(html_data,"lxml")
    content=soup.find('div',{'class':'x6s0dn4 x78zum5 xdt5ytf x5yr21d x1odjw0f x1n2onr6 xh8yej3'})

    all_span=content.find_all('span',{'class':'x1lliihq x1plvlek xryxfnj x1n2onr6 x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs x1s928wv xhkezso x1gmr53x x1cpjm7i x1fgarty x1943h6x x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj'})
    for name in all_span:
        if name.string[0]!="#":
            account_names.append(name.string)

    pop_search=driver.find_element('xpath','/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[1]/div/div/div[2]')
    pop_search.click()
    
    return account_names
```

```python
toSearch='food'
names=search(toSearch)
for name in names:
    print(name)
```

### 3.Searching and opening a profile

```python
def open_profile(handle_name):
    #Search field find and clicked
    search_field = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div/a/div')))
    search_field.click()
    #given account handle searched in searchbox
    search_box = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[1]/div/div/input')))
    search_box.send_keys({handle_name})
    time.sleep(2)
    #Handle name searching and opening
    id= wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div[2]/div/div/div[3]/div/a[1]/div[1]/div/div/div[2]/div/div/div/span")))
    if id.text == handle_name:
        id.click()
    else:
        return f"The Account {handle_name} is not visible."
    
    return f"The Profile {handle_name} has been searched successful"
```

```python
handle_name = "sodelhi"
response= open_profile(handle_name)
print(response)
```

### 4.follow and Unfollow given handle
1. Open the Handle
2. Start following it. print a message if you already following.
3. After following unfollow the handle. print a message if already unfollowed

```python
def follow_the_handle():
    try:    
        follow_button = wait.until(EC.presence_of_element_located((By.TAG_NAME,"button")))
        if follow_button.text != "Following":
            follow_button.click()
            print(f"you are now following {handle_name}")
        else:
            print(f"You already following this account {handle_name}")
    except TimeoutException:
        print("There is some problem exists in the code")
```

```python
def unfollow_the_handle():
    try:
        unfollow_button = wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/div[1]/div[1]/div/div[1]/button/div/div[1]")))
        if unfollow_button.text=='Following':
            unfollow_button.click()
            time.sleep(2)
            confirm_unfollow_button= wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[7]/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div/div[8]")))
            confirm_unfollow_button.click()
            print(f"you are now unfollowed {handle_name}.")
        else:
            print(f"You are not following {handle_name}.")
    except TimeoutException:
        print("There is some problem exists in the code")
```

### 5. Like/Unlike Posts
1. like top 30 posts of the given handle.print a message if you already like it.
2. Unlike top 30 posts of the given handle.print a message if you already unliked it.

```python
def like_post():
    #Scrolling the webpage for more content
    # page_height = driver.execute_script("return document.body.scrollHeight")
    driver.execute_script(f"window.scrollTo(0,6000)")
    time.sleep(5)
    driver.execute_script(f"window.scrollTo(0,-6000)")
    
    
    #locating post
    posts = driver.find_elements(By.CLASS_NAME,'_aagu')
    for i in range(30):
        posts[i].click()
        time.sleep(2)
        like = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[8]/div[1]/div/div[3]/div/div/div/div/div[2]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/div/div')))
        st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
        if st=="Like":
            like.click()
            print(f"You liked the post number: {i+1}")
            time.sleep(2)
        else:
            print('You have already LIKED Post Number :', i+1)
            time.sleep(2)
            
        #locating cross button for closing post
        driver.find_element(By.XPATH,"/html/body/div[8]/div[1]/div/div[2]/div/div").click()
```

```python
def unlike_post():
    #Scrolling the webpage for more content
    # page_height = driver.execute_script("return document.body.scrollHeight")
    driver.execute_script(f"window.scrollTo(0,6000)")
    time.sleep(5)
    driver.execute_script(f"window.scrollTo(0,-6000)")
    time.sleep(2)
    
    
    posts = driver.find_elements(By.CLASS_NAME,'_aagu')
    for i in range(30):
        posts[i].click()
        time.sleep(2)
        unlike=wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[8]/div[1]/div/div[3]/div/div/div/div/div[2]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/div/div')))
        st = BeautifulSoup(unlike.get_attribute('innerHTML'),"html.parser").svg['aria-label']
        if st=="Unlike":
            unlike.click()
            print(f"You unliked the post number: {i+1}")
            time.sleep(2)
        else:
            print('You have already unliked Post Number :', i+1)
            time.sleep(2)
    
        #locating cross button for closing post
        driver.find_element(By.XPATH,"/html/body/div[8]/div[1]/div/div[2]/div/div").click()
```

### 6.Extract list of followers
1. extract the usernames of first 500 followers.
2. print all the followers of 'so delhi' that you are following but those who don't follow you back.

```python
def Extract_Followers():
    
    # locating followers button and click on it
    follower_btn=driver.find_element(By.XPATH,"/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]/a")
    follower_btn.click()
    time.sleep(2)
        
    #locating followers list
    frame = driver.find_element(By.CLASS_NAME,'_aano')
        
    #scrolling untill first 500 user is located.
    for i in range(50):
        time.sleep(2)
        driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",frame)
        
    names=[]
    #extracting userdata
    followers = frame.find_elements(By.CLASS_NAME,'_aacx')
    #appending all 500 usernames  to a list
    for i in followers[:500]:
        names.append(i.text)
    
    #close the dialog box
    driver.find_element(By.CLASS_NAME,"_abm0").click()
    return names
    
```

#####  Print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def Following():
    try :
        # locating following button and click on it
        followers_btn = wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[3]/a')))
        followers_btn.click()
        time.sleep(2)
        
        #locating followers list
        frame = driver.find_element('xpath','/html/body/div[6]/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[4]')
        for i in range(20):
            time.sleep(1)
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",frame)
        #Extracting all the following lists
        names = []
        following = driver.find_elements('class name','_aad7')
        for i in following:
            names.append(i.text)
        
        driver.find_element(By.CLASS_NAME,"_abm0").click()
        
        return names
        
        
    
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
#search and open 'foodtalkindia' instagram handle
open_profile('foodtalkindia')
# Extracting followers using Extract_Followers() function
time.sleep(3)
followers_of_foodtalkindia = Extract_Followers()
#casting into set
followers_of_foodtalkindia = set(followers_of_foodtalkindia)
```

```python
#now finding all user followed by me.
open_profile("atharva_hatekar")
time.sleep(3)
followed_by_me = Following()
followed_by_me = set(followed_by_me)
```

```python
#taking intersection so s1 contains only that user who followed by me
s1=(followers_of_foodtalkindia).intersection(followed_by_me)
if len(s1) == 0:
    print('No such users found')
else:
    #now extracting my followers
    my_follower = Extract_Followers()
    my_follower = set(my_follower)
    #taking intersection with s1, so s2 contains only users that I am following them but they don’t follow me
    s2=s1.intersection(my_follower)
    if len(s2) == 0:
        print('No such users found')
    else:
        for user in s2:
            print(user)
```

### Problem 7 : Check the story of ‘coding.ninjas’
Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -  
1.If You have already seen the story.  
2.Or The user has no story.  
3.Or View the story if not yet seen.  


```python
def Check_Story():
    try:
        #locating story or profile pic
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aarg")))
        
        #check the Profile photo size to find out story is available or not
        height = driver.find_element(By.CLASS_NAME,"_aarh").get_attribute('height')
        if int(height) == 166:
            print("Already seen the story")
        else:
            print("Viewing the story")
            driver.execute_script("arguments[0].click();",story)
    
    except TimeoutException:
        print("No story is available to view")
```
